```python
def k_nearest_roads(latitud, longitud, gdf, k=5, radi=None):
    punto = Point(longitud, latitud)

    gdf = gdf.to_crs(epsg=25831)
    punto_utm = gpd.GeoSeries([punto], crs="EPSG:4326").to_crs(epsg=25831).iloc[0]

    gdf = gdf[gdf.geometry.notnull() & gdf.is_valid & ~gdf.geometry.is_empty]

    gdf['distance'] = gdf.geometry.distance(punto_utm)

    if radi is not None:
        gdf = gdf[gdf['distance'] <= radi]

    k_cercanas = gdf.nsmallest(k, 'distance')

    return k_cercanas

shapefile_path = "../Data/OpenTransportMap/roadlinks_ES511.shp"
gdf = gpd.read_file(shapefile_path)

latitud = 41.3851
longitud = 2.1734
radi = 1000

carreteras_cercanas = k_nearest_roads(latitud, longitud, gdf, k=5, radi=radi)

print(carreteras_cercanas[['roadname', 'functional', 'formofway', 'distance']])
```

```python
# Coordenades de latitud i longitud donades
latitud = 41.3851
longitud = 2.1734
punt = Point(longitud, latitud)

# Reprojectar el GeoDataFrame a EPSG:25831 (UTM zona 31N)
gdf = gdf.to_crs(epsg=25831)

# Reprojectar el punt a EPSG:25831
punt_utm = gpd.GeoSeries([punt], crs="EPSG:4326").to_crs(epsg=25831).iloc[0]

# Calcular la distància en metres
gdf['distance'] = gdf.geometry.distance(punt_utm)

# Trobar la carretera més propera
carretera_mes_propera = gdf.loc[gdf['distance'].idxmin()]

# Mostrar informació de la carretera més propera
print(carretera_mes_propera[['roadname', 'functional', 'formofway', 'distance']])
```
